```python
import pandas as pd
import numpy as np
```

```python

# load data
col_name = ['Danh_muc', 'Ten_san_pham', 'Ton_co_the_ban', 'Gia_ban', 'So_ban_hom_qua', 'So_ban_3_ngay_qua', 'So_ban_7_ngay_qua', 'So_nhap_hom_qua', 'So_nhap_7_ngay_qua']
data = pd.read_excel(r'/Users/loinguyen/Documents/CV-Interview/CellPhoneS_DA_test/Test_BI_1.xlsx', sheet_name='DATA 1 - TỒN KHO THÁNG 7')
df = pd.DataFrame(data)
df.head()

# crate danh mục 2 include only brand 

new = df['Danh mục'].str.split(" ", 
                                      expand = False, 
                                      n  = 1)
new_list = new.to_list()
df_new = pd.DataFrame(new_list, columns = ['Danh mục 2', 'Tình trạng'])
df['Danh mục 2'] = df_new['Danh mục 2']
df['Tình trạng'] = df_new['Tình trạng']

# create condition for check 

df['check'] = ''
for i in range(0, df.shape[0]):
    sub_txt = df['Danh mục 2'][i]
    txt = df['Tên sản phẩm'][i].upper()
    df['check'][i] = txt.find(sub_txt)

# unique danh mục sp
danh_muc = df['Danh mục 2'].unique().tolist()
danh_muc.remove('KHÁC')
danh_muc.remove('APPLE')

# checking 
for i in range(0, df.shape[0]):
    if (df['check'][i]== -1):
        if (df['Danh mục 2'][i] == 'KHÁC'):
            txt = df['Tên sản phẩm'][i].upper()   
            for item in danh_muc:
                if txt.find(item) != -1:
                    df['Danh mục'][i] = ((item+" CŨ"))
        else:
            txt = df['Tên sản phẩm'][i].upper()   
            for item in danh_muc:
                if txt.find(item) != -1:
                    df['Danh mục'][i] = ((item+" CŨ"))

#drop unesserary columns

df = df.drop(['Danh mục 2', 'Tình trạng', 'check'], axis = 1)

#export excel file
df.to_excel("df_ton_kho_7.xlsx", index = False)
```
